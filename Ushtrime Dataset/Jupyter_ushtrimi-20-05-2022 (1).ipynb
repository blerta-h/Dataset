```python
#E kemi csv file dhe te njejten do ta lexoni 
#Duheni ta paraqisni vizualisht ne njeren nga stilet qe ju mendoni 
#-per te paraqitur grafin
#Pasi kemi titullin eligjeratave sot subplots atehere le te miren tri 
#-parametra
#style: seaborn
#parametrat age, All_Devs, Python, JavaScript

import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')
data = pd.read_csv('Data-20-05-2022.csv')
mosha= data['Age']
paga_zhv= data['All_Devs']
python= data['Python']
javascript= data['JavaScript']

fig,(ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(mosha, paga_zhv, linestyle='--', color= 'green', label='Te gjithe zhvilluesit')
ax2.plot(mosha, python, label='Python')
ax2.plot(mosha, javascript, label= 'Javascript')


ax1.legend()

ax1.set_title('Mediana e pagave ne baze te moshes')
ax1.set_xlabel('Mosha')
ax1.set_ylabel('Mediana e pagave')

ax2.legend()

ax2.set_title('Mediana e pagave ne baze te moshes')
ax2.set_xlabel('Mosha')
ax2.set_ylabel('Mediana e pagave')

plt.show()
```
